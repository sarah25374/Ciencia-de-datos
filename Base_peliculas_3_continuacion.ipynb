```python
fig = px.box(df, y='index', title='Deteccion de datos atipicos')
fig.update_layout(width=300,height=500)
fig.show()
```

```python
iqr=df['Año de lanzamiento'].quantile(0.75)-df['Año de lanzamiento'].quantile(0.25)
filtro_inferior = df['Año de lanzamiento'] > (df['Año de lanzamiento'].quantile(0.25) - (iqr * 1.5))
filtro_superior = df['Año de lanzamiento'] < (df['Año de lanzamiento'].quantile(0.75) + (iqr * 1.3))
df_filtrado= df[filtro_inferior & filtro_superior]
fig = px.box(df_filtrado, y='Año de lanzamiento', title='Deteccion de datos atipicos')
fig.update_layout(width=300,height=500)
fig.show()
```

```python
iqr=df['Duracion'].quantile(0.75)-df['Duracion'].quantile(0.25)
filtro_superior = df['Duracion'] < (df['Duracion'].quantile(0.75) + (iqr * 1.5))
filtro_inferior = df['Duracion'] > (df['Duracion'].quantile(0.25) - (iqr * 1.3))
df_filtrado= df[filtro_inferior & filtro_superior]
fig = px.box(df_filtrado, y='Duracion', title='Deteccion de datos atipicos')
fig.update_layout(width=300,height=500)
fig.show()
```

```python
iqr=df['imdb_puntuacion'].quantile(0.75)-df['imdb_puntuacion'].quantile(0.25)
filtro_superior = df['imdb_puntuacion'] < (df['imdb_puntuacion'].quantile(0.75) + (iqr * 1.5))
filtro_inferior = df['imdb_puntuacion'] > (df['imdb_puntuacion'].quantile(0.25) - (iqr * 1.3))
df_filtrado= df[filtro_inferior & filtro_superior]
fig = px.box(df_filtrado, y='imdb_puntuacion', title='Deteccion de datos atipicos')
fig.update_layout(width=300,height=500)
fig= px.box(df, y='imdb_puntuacion', title=' Datos atipicos eliminados')
fig.update_layout(width=300,height=500)
fig.show()
```

```python
iqr=df['imdb_votos'].quantile(0.75)-df['imdb_votos'].quantile(0.25)
filtro_superior = df['imdb_votos'] < (df['imdb_votos'].quantile(0.75) + (iqr * 1.5))
filtro_inferior = df['imdb_votos'] > (df['imdb_votos'].quantile(0.25) - (iqr * 1.3))
df_filtrado= df[filtro_inferior & filtro_superior]
fig = px.box(df_filtrado, y='imdb_votos', title='Deteccion de datos atipicos')
fig.update_layout(width=300,height=500)
fig.show()
fig= px.box(df, y='imdb_votos', title=' Datos atipicos eliminados')
fig.update_layout(width=300,height=500)
fig.show()
```
